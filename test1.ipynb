```python
import pandas as pd
```

```python
#1、读取文件ForecastingOrders.csv，。(2分)
df=pd.read_csv("data/ForecastingOrders.csv")
df
```

```python
#查看数据的行列数和前10行数据
df.shape
```

```python
df.head(10)
```

```python
#2、将数据进行去重。(2分)
df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
max=df["Order type A"].max()
min=df["Order type A"].min()
pjz=(max+min)/2
pjz
```

```python
#3、将“Order type A”字段为空的数据，填充为该列最大值和最小值的平均值。(3分)
df["Order type A"]=df["Order type A"].fillna(pjz)
df["Order type A"]
```

```python
#4、划分数据，形成训练集和测试集。(3分)
x=df.iloc[:,:-1]
x
```

```python
y=df["Target (Total orders)"]
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#Xgboost回归
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV
xg=GradientBoostingRegressor(loss="squared_error",learning_rate=0.1,n_estimators=100)
#对以上模型分别进行网格搜索调优。(3分)
xg_parameters = {'loss':['huber','ls'], 'learning_rate':[0.1, 0.3],'subsample':[1.0,3.0]}
xg_model=GridSearchCV(xg,param_grid=xg_parameters,cv=5)
#训练
xg_model.fit(x_train,y_train)
```

```python
#最佳模型
best_xg_model=xg_model.best_estimator_
best_xg_model
```

```python
#回归的评估
#8、模型评估，使用MAE、MSE指标进行评估
from sklearn.metrics import mean_absolute_error,mean_squared_error
best_xg_model_pre=best_xg_model.predict(x_test)
mean_absolute_error(y_test,best_xg_model_pre)
mean_squared_error(y_test,best_xg_model_pre)
```

```python
#随机森林回归
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=100,max_depth=100,min_samples_split=2,)
#网格调优
rf_parameters = {'min_samples_split':[2,4], 'n_estimators':[10, 20],'max_depth':[100,200]}
rf_model=GridSearchCV(rf,param_grid=rf_parameters,cv=5)
#训练
rf_model.fit(x_train,y_train)
```

```python
#随机森林最佳模型
best_rf_model=rf_model.best_estimator_
best_rf_model
```

```python
#随机森林模型评估
from sklearn.metrics import mean_absolute_error,mean_squared_error
best_rf_model_pre=best_rf_model.predict(x_test)
mean_absolute_error(y_test,best_rf_model_pre)
mean_squared_error(y_test,best_rf_model_pre)
```

```python
#9、输出两个模型的预测结果。(3分)
#Xgboost预测结果
best_xg_model_pre=best_xg_model.predict(x_test)
best_xg_model_pre
```

```python
#随机森林预测结果
best_rf_model_pre=best_rf_model.predict(x_test)
best_rf_model_pre
```
