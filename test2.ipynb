```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import GridSearchCV
```

```python
#1、从名为 “SouthGermanCredit.csv” 的文件中读取数据。（2 分）
df=pd.read_csv("data/SouthGermanCredit.csv")
df
```

```python
#6、使用label编码或者独热编码或者其他方式，将 “gender” 列中的数据转换为数值形式。（4 分）
df['gender']=df['gender'].replace({'f':0,'m':1})
df
```

```python
#3、若数据里存在缺失值的行，直接删除这些行。（2 分）
df.isnull().sum()
df.dropna()
```

```python
#4、查看数据是否有重复行，若存在则删除重复行。（2 分）
df.duplicated()
df.drop_duplicates(inplace=True)
```

```python
xx=df[['hoehe']]
xx
```

```python
#5、对数据中名为 “hoehe” 的列进行标准化处理。（4 分）
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ss.fit_transform(df[['hoehe']])
ss.transform(xx)
```

```python
#7、利用卡方检验来进行特征选择，选出最优的 10 个特征。（4 分）
x=df.iloc[:,:-1]
x
```

```python
y=df['kredit']
y
```

```python
#8、把数据集按照 80% 与 20% 的比例划分为训练集和测试集。（4 分）
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#逻辑回归
#l1
lr1=LogisticRegression(penalty="l1",solver='liblinear')
#网格搜索
#参数至少设置两个
lr1_parameters = {'intercept_scaling':[1,3], 'max_iter':[100, 200]}
lr1.model=GridSearchCV(lr1,param_grid=lr1_parameters,cv=5)
#训练
lr1.model.fit(x_train,y_train)
```

```python
#最佳模型l1
best_lr1_model=lr1.model.best_estimator_
best_lr1_model
```

```python
#逻辑回归l2
lr2=LogisticRegression(penalty="l2")
#网格搜索
#参数至少设置两个
lr2_parameters = {'intercept_scaling':[1,3], 'max_iter':[100, 200]}
lr2.model=GridSearchCV(lr2,param_grid=lr2_parameters,cv=5)
#训练
lr2.model.fit(x_train,y_train)
```

```python
#最佳模型l2
best_lr2_model=lr2.model.best_estimator_
best_lr2_model
```

```python
#并分别打印出它们的截距和系数。（4 分）
print(f'l1截距:{best_lr1_model.intercept_}')
print(f'l1系数:{best_lr1_model.coef_}')
print(f'l2截距:{best_lr2_model.intercept_}')
print(f'l2系数:{best_lr2_model.coef_}')
```

```python
#12、选用至少两种评价指标来评估模型性能。（4 分）
#l1评估
from sklearn.metrics import accuracy_score,recall_score
best_lr1_model_pre=best_lr1_model.predict(x_test)
l1_acc = accuracy_score(y_test, best_lr1_model_pre)
recall_score(y_test,best_lr1_model_pre)
```

```python
#l2评估
best_lr2_model_pre=best_lr2_model.predict(x_test)
l2_acc = accuracy_score(y_test, best_lr2_model_pre)
recall_score(y_test,best_lr2_model_pre)
```

```python
#13、保存性能较好的那个最优模型。（4 分）
import joblib
if l1_acc>l2_acc:
    joblib.dump(filename='best_lr1_model.pkl',value=True)
else:
    joblib.dump(filename='best_lr2_model.pkl',value=True)
```
